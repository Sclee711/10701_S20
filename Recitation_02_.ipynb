# Entropy and Information Gain

We saw in class that information gain can be used to determine which variable to split on while constructing a decision tree.

### Entropy

The entropy of a (discrete) random variable X taking values $V_1...V_m$ is defined as:

$$H(X)=-\sum_{i=1}^m P(X=V_i)\log_2(P(X=V_i))$$

Entropy is a function of the probability distribution of X. If X is uniform, the entropy will be large (we have little information about the value of X):

```python
Values_X = np.array(['a','b','c','d'])
P_X = np.array([1/4,1/4,1/4,1/4])

plt.bar(np.arange(4), P_X, tick_label = Values_X)
plt.ylabel('P(X)'); plt.xlabel('X'); plt.ylim([0,1])

H_X = -np.sum(P_X*np.log2(P_X)) # python * is element wise, use np.dot for matrix multiplication

print('H(X) = {}'.format(H_X))
```

If X is not uniform, and instead has some very likely and very unlikely values, entropy will be low (we have more information about X):

```python
Values_X = np.array(['a','b','c','d'])
P_X = np.array([1/8,1/2,1/8,1/4])

plt.bar(np.arange(4), P_X, tick_label = Values_X)
plt.ylabel('P(X)'); plt.xlabel('X'); plt.ylim([0,1])

H_X = -np.sum(P_X*np.log2(P_X)) # python * is element wise, use np.dot for matrix multiplication

print('H(X) = {}'.format(H_X))
```

Think of encoding the values of X in the second case using the following code on a serial channel:

| X    | a   | b   | c   | d   |
|------|-----|-----|-----|-----|
| P(X) | 1/8 | 1/2 | 1/8 | 1/4 |
| code(X) | 00 | 01 | 10 | 11 |
| N(X) (bits) | 2 | 2 | 2 | 2 |

The expected value of the number of bits to use $N$ is:
$$E[N] = \sum_{i=1}^4 N_i\times P(X=V_i) = 2\times1/8 + 2\times1/2 + 2\times1/8 + 2\times1/4 = 2$$

Consider the code:

| X    | a   | b   | c   | d   |
|------|-----|-----|-----|-----|
| P(X) | 1/8 | 1/2 | 1/8 | 1/4 |
| code(X) | 110 | 0 | 111 | 10 |
| N(X) (bits) | 3 | 1 | 3 | 2 |

The expected value for the number of bits:

$$E[N] = \sum_{i=1}^4 N_i\times P(X=V_i) = 3\times1/8 + 1\times1/2 + 3\times1/8 + 2\times1/4 = 1.75$$

Entropy corresponds to the smallest number of bits on average per symbol needed to transmit symbols on a serial channel. 


### Entropy of coin flip example

As an example, here is how the entropy of a coin flip varies as the probability of heads vary. When the probability is 0.5, we have the least information about the outcome (highest entropy).

## Information Gain

How much would knowing about X help me? (how many bits would I save?)

$$\text{IG}(Y|X) = H(Y) - H(Y|X)$$

In the previous example, 4 students walked to school and 4 did not:

$$ H(Y) = -(0.5\times\log_2(5)+0.5\times\log_2(5)) = 1$$

$$\text{IG}(Y|X) = H(Y) - H(Y|X) = 1- 0.5 =0.5$$

We use Information Gain to find features to construct a decision tree.

## In class example

```python
X_names = np.array(['Outlook','Temperature','Humidity','Wind'])

Weather = np.array([['Sunny','Hot','High','Weak'],
                    ['Sunny','Hot','High','Strong'],
                    ['Overcast','Hot','High','Weak'],
                    ['Rain','Mild','High','Weak'],
                    ['Rain','Cool','Normal','Weak'],
                    ['Rain','Cool','Normal','Strong'],
                    ['Overcast','Cool','Normal','Strong'],
                    ['Sunny','Mild','High','Weak'],
                    ['Sunny','Cool','Normal','Weak'],
                    ['Rain','Mild','Normal','Weak'],
                    ['Sunny','Mild','Normal','Strong'],
                    ['Overcast','Mild','High','Strong'],
                    ['Overcast','Hot','Normal','Weak'],
                    ['Rain','Mild','High','Strong']])

PlayTennis = np.array([['No'],['No'],['Yes'],['Yes'],['Yes'],
                       ['No'],['Yes'],['No'],['Yes'],['Yes'],
                       ['Yes'],['Yes'],['Yes'],['No']])

def entropy(vec):
    # Computes entropy of a vector of discrete variables
    values = np.unique(vec)
    p_X = np.array([np.mean(vec==v) for v in values])
    return -np.sum(p_X*np.log2(p_X))
    
H_Y = entropy(PlayTennis)
print(H_Y)
```

Compute $\text{IG}(Y|X)$ for every X:

```python
IG = np.zeros(len(X_names))
for i_var, X_var in enumerate(X_names):
    values = np.unique(Weather[:,i_var])
    n_values = len(values)
    H_YX = np.zeros(n_values)
    f, axs = plt.subplots(1, n_values+1, sharey=True, figsize = (10,0.5))
    for i_ax,ax in enumerate(axs[:-1]):
        value = values[i_ax]
        # plot histogram
        PT = PlayTennis[Weather[:,i_var]==value]
        ax.hist(PT); ax.set_title(value); ax.set_ylim([0,5])
        # compute H(Y|X=V_i)xP(X=V_i)
        P_V_i = np.mean(Weather[:,i_var]==value)
        H_YX[i_ax] = entropy(PT)*P_V_i        
    axs[-1].set_title(X_var); axs[-1].axis('off')
    IG[i_var] = H_Y - np.sum(H_YX)
```

## Information Gain, first level:

```python
plt.bar(np.arange(len(X_names)),IG)
plt.xticks(ticks=np.arange(len(X_names)),labels=X_names,rotation=45);
plt.ylabel('IG');
```
